```python
import numpy as np
import pandas as pd
```

```python
all_sit = pd.read_csv('2019_2020_team_stats.csv'); all_sit.index += 1
all_sit = all_sit[ ['Tm', 'PDO', 'FF%'] ]
all_sit['Test'] = all_sit.index
all_sit['Act_Rk'] = [1, 4, 3, 18, 2, 17, 12, 20, 6, 23, 22, 5, 7, 13, 10, 11, 21, 15, 9, 19, 25, 16, 14, \
                     8, 26, 30, 27, 24, 29, 28, 31]
all_sit.head()
```

```python
pp = pd.read_csv('2019_2020_team_stats_pp.csv'); pp.index += 1
pp = pp[ ['Tm',  'FF%'] ]
pp.rename(columns={'FF%': 'FF%pp'}, inplace=True)
pp.head()
```

```python
allSitPDO_FFpercPp = pd.merge(all_sit, pp, on='Tm'); allSitPDO_FFpercPp.index +=1
allSitPDO_FFpercPp.head()
```

```python
func = lambda row: row['PDO'] + 0.3*row['FF%'] + 0.1*row['FF%pp']

allSitPDO_FFpercPp['NEW'] = allSitPDO_FFpercPp.apply(func, axis=1)
allSitPDO_FFpercPp.index += 1

new_df = allSitPDO_FFpercPp[ [ 'Act_Rk', 'Tm', 'NEW'] ].copy()
sorted_df = new_df.sort_values(by=['NEW'], ascending=False).reset_index(drop=True)
sorted_df['Rank'] = sorted_df.index
sorted_df['Diff'] = np.abs( sorted_df['Act_Rk'] - sorted_df.index )
sorted_df = sorted_df[ ['Tm', 'NEW', 'Rank', 'Act_Rk', 'Diff'] ] 
sorted_df
```
